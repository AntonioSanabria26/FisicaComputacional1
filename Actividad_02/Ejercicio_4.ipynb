# **Aproximación de la función ln(1+x)**

Aproximación de grado 4 \\

$$ x-x^2/2+x^3/3-x^4/4 $$

Aproximación de grado 7 \\

$$ x-x^2/2+x^3/3-x^4/4+x^5/5-x^6/6+x^7/7 $$
 
Aproximación de orden 11 \\

$$ x-x^2/2+x^3/3-x^4/4+x^5/5-x^6/6+x^7/7-x^8/8+x^9/9-x^{10}/10+x^{11}/11 $$

Aproximación de orden 16 \\

$$ x-x^2/2+x^3/3-x^4/4+x^5/5-x^6/6+x^7/7-x^8/8+x^9/9-x^{10}/10+x^{11}/11-x^{12}/12+x^{13}/13-x^{14}/14+x^{15}/15-x^{16}/16 $$

```
#---------------BIBLIOTECAS---------------#
import matplotlib.pyplot as plt
import numpy as np
import math

#--------- VARIABLES Y FUNCIONES ---------#
x = np.linspace(-1.5,1.5,100)  # arrego de 100 puntos entre -1.5 a 1.5
y = np.log(1+x)

sT4 = x-x**2/2+x**3/3-x**4/4

sT7 = x-x**2/2+x**3/3-x**4/4+x**5/5-x**6/6+x**7/7

sT11 =  (x-x**2/2+x**3/3-x**4/4+x**5/5-x**6/6+x**7/7
          -x**8/8+x**9/9-x**(10)/10+x**(11)/11)

sT16 = (x-x**2/2+x**3/3-x**4/4+x**5/5-x**6/6+x**7/7-x**8/8+x**9/9-x**(10)/10
        +x**(11)/11-x**(12)/12+x**(13)/13-x**(14)/14+x**(15)/15-x**(16)/16)

#---------- CREACiÓN DE LA GRÁFICA ---------#
plt.subplot(111)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Aproximación de taylor a ln(1+X)")

#---------- GRAFICANDO FUNCIONES -----------#
plt.plot(x,sT4, label='Serie Taylor Orden 4')
plt.plot(x,sT7, label='Serie Taylor Orden 7')
plt.plot(x,sT11, label='Serie Taylor Orden 11')
plt.plot(x,sT16, label='Serie Taylor Orden 16')
plt.plot(x,y, label='ln(1+x)')

# Agregar una malla
plt.grid(True)
# Fijar los límites en el eje Y.
plt.ylim(-4, 2)
# Ubicación de las etiquetas de las funciones
plt.legend(bbox_to_anchor=(1,1), loc="upper left")
plt.show()

```
